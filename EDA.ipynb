```python
# Import libraries
!pip install pandas seaborn matplotlib
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("train.csv")
```

```python
# Basic Data Exploration

print("=== Dataset Info ===")
print(df.info())

print("\n=== Dataset Description ===")
print(df.describe())
```

```python
# Value counts for categorical columns
print("\n=== Value Counts for Categorical Columns ===")
for col in df.select_dtypes(include=['object', 'category']).columns:
    print(f"\nColumn: {col}")
    print(df[col].value_counts())
```

```python
# Pairplot
sns.pairplot(df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']])
plt.suptitle("Pairplot of Numeric Features", y=1.02)
plt.show()
```

```python
# Correlation Heatmap

plt.figure(figsize=(8, 6))
sns.heatmap(df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].corr(),
            annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Histograms

df[['Age', 'Fare', 'SibSp', 'Parch']].hist(
    bins=20, figsize=(10, 8), color='skyblue', edgecolor='black')
plt.suptitle("Histograms of Numeric Features")
plt.show()
```

```python
# Boxplots

plt.figure(figsize=(10, 6))
sns.boxplot(x='Survived', y='Age', data=df)
plt.title("Boxplot of Age vs Survival")
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title("Boxplot of Fare vs Survival")
plt.show()
```

```python
# Scatterplot

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df, palette='coolwarm')
plt.title("Scatterplot of Age vs Fare (Colored by Survival)")
plt.show()
```

```python
print("""
OBSERVATIONS:
1. Higher survival rate in Pclass 1 and among younger passengers.
2. Fare is positively correlated with survival.
3. Most passengers are between 20–40 years old.
4. Many traveled alone (SibSp=0, Parch=0).
5. Higher class (lower Pclass number) = higher fare and better survival.
""")
```
